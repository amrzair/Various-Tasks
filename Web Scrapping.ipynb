## Scrap the website csimarket.com to get the value of a specific kpi and keep it updated.

# Import required libraries

```python
import requests
from bs4 import BeautifulSoup
```

# showing the web data as text

```python
csi_url = "https://csimarket.com/Industry/industry_ManagementEffectiveness.php?ind=1101"
response = requests.request("GET",csi_url)
print(response.text)
```

# Showing the data as html to be read by python

```python
soup = BeautifulSoup(response.text, "html.parser")
print(soup)
```

# Targeting the required KPI 
### ROA (Return On Asset)

```python
data = soup.find_all("td", attrs={"class":"sivirub strong"})
print(data)
```

# Targetting the required value
### showint the ROA (Return On Asset) KPI at 2021

```python
data_a = str(data)
roa_kpi_2021 = data_a[61:66]
print(roa_kpi_2021)
```
